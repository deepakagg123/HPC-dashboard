## Reading Accounting Files

```python
 from glob import glob
files = glob('data/PBS_accounting_logs/202307*')
files.sort()
print(files)
# Python program to
# demonstrate merging of
# two files

# Creating a list of filenames
filenames = files

# Open file3 in write mode
with open('file3.txt', 'w') as outfile:

    # Iterate through list
    for names in filenames:

        # Open each file in read mode
        with open(names) as infile:

            # read the data from file1 and
            # file2 and write it in file3
            outfile.write(infile.read())

        # Add '\n' to enter data of file2
        # from next line
        outfile.write("\n")
```

## Making Dataframe

```python
data_dict = dict()
data_dict['log_date'] = []
data_dict['State'] = []
data_dict['job_id'] = []
with open("data1.text", 'r') as data_file:
    content = data_file.readlines();
#     print(data_file.readlines(33))

    for line in content:
        parts = line.split(';')
        data_dict['log_date'].append(parts[0])
        data_dict['State'].append(parts[1])
        data_dict['job_id'].append(parts[2])
        
        # print(parts[3])
        for info in parts[3].split(' '):
            # print(info)
#             for key,value in info.split('=')  :
#             key,value = key.split(':')
            # print(type(info))
            cond1 = info.find("exec_vnode")
            cond2 = info.find("Resource_List.select")
            cond3 = info.find("=")
            if cond1 == -1 and cond2 == -1 and cond3 != -1 :
                key, value = info.split('=')
#                 print('key=',key)
#                 print('value=',value)

                if key in data_dict.keys():
                        data_dict[key].append(value)
                else:
                        data_dict[key] = [value]
```

```python
df1 = pd.DataFrame.from_dict(data_dict, orient ='index')
#df1
```

```python
max_jobs_by_user = df3.groupby('user')['job_id'].count().reset_index()
#max_jobs_by_user
max_jobs_by_user.columns = ['user', 'max_jobs_submitted']
max_jobs_by_user
```

```python
df3['Resource_List.ncpus'] = df3['Resource_List.ncpus'].astype(int)
total_ncpus_by_user = df3.groupby('user')['Resource_List.ncpus'].sum().reset_index()
total_ncpus_by_user.columns = ['user', 'total_ncpus_used']
total_ncpus_by_user
```

```python
df3.dropna(subset=['Resource_List.ngpus'], inplace=True)
df3['Resource_List.ngpus'] = df3['Resource_List.ngpus'].astype(int)
total_ngpus_by_user = df3.groupby('user')['Resource_List.ngpus'].sum().reset_index()
total_ngpus_by_user.columns = ['user', 'total_ngpus_used']
total_ngpus_by_user
#df3
```

```python
df3['resources_used.walltime'] = pd.to_timedelta(df3['resources_used.walltime'])
#df3

# Calculate the total walltime used (in seconds) by each user for all jobs submitted
df3['total_walltime_used'] = df3.groupby('user')['resources_used.walltime'].transform('sum').dt.total_seconds()

# Drop any rows with NaN values (if any) in the new column
df3.dropna(subset=['total_walltime_used'], inplace=True)
df3
df3['resources_used.walltime'] = df3['resources_used.walltime'].astype(int)
total_walltime_by_user = df3.groupby('user')['resources_used.walltime'].sum().reset_index()
total_walltime_by_user.columns = ['user', 'total_ngpus_used']
total_walltime_by_user
```
