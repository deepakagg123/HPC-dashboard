```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from glob import glob
y = " "

def searchData3():
    s1 = 'Resource_List.ngpus='
    obj1=open("data1.text","r")

    for line in obj1.readlines():
        if s1 in line:
            global y
            y += line
    obj1.close()
searchData3()
outfile1 = open('gpus.text','w')
outfile1.write(y)
outfile1.close()
```

```python
data_dict = dict()
data_dict['log_date'] = []
data_dict['State'] = []
data_dict['job_id'] = []
with open("gpus.text", 'r') as data_file:
    content = data_file.readlines();
#     print(data_file.readlines(33))

    for line in content:
        parts = line.split(';')
        data_dict['log_date'].append(parts[0])
        data_dict['State'].append(parts[1])
        data_dict['job_id'].append(parts[2])
        
#         print("Part0--->{} , part1--->{} ,part2--->{} ,part3--->{}".format(parts[0],parts[1],parts[2],parts[3]))
        for info in parts[3].split(' '):
            # print(info)
#             for key,value in info.split('=')  :
#             key,value = key.split(':')
            # print(type(info))
            cond1 = info.find("exec_vnode")
            cond2 = info.find("Resource_List.select")
            cond3 = info.find("=")
            if cond1 == -1 and cond2 == -1 and cond3 != -1 :
                key, value = info.split('=')
#                 print('key=',key)
#                 print('value=',value)

                if key in data_dict.keys():
                        data_dict[key].append(value)
                else:
                        data_dict[key] = [value]
```

```python
df = pd.DataFrame.from_dict(data_dict, orient ='index')
gpu=df.transpose()
gpu
```

```python
gpu.drop(columns=['group','project','jobname','ctime','qtime','etime','start','Resource_List.nodect','Resource_List.mpiprocs','exec_host','resources_used.vmem','run_count','Resource_List.place','resources_used.mem','end','session',
'resource_assigned.mem','Resource_List.arch'], inplace=True)
```

```python
gpu['Resource_List.ngpus'] = gpu['Resource_List.ngpus'].astype(float)
```
